# Emacs

## What exactly is Emacs?

Emacs is a text editing tool that comes pre-installed with Linux and macOS. 

Its purpose is to provide a versatile environment for ***editing text, writing code, and performing various tasks related to text manipulation and customization.***

It was developed by Richard Stallman and has a long history within the free software and open-source communities.

### Emacs Terms You Need To Know
1. Buffer: The text you are editing in Emacs resides in an object called a buffer. Each time you visit a file, a buffer is used to hold the file’s text. ***Think of a buffer as tabs in a web browser***
    - A buffer is an area or container that holds text and represents a file or a document being edited. A buffer is where you can view, edit, and manipulate the contents of a file or perform other tasks within Emacs.

## Basic Emacs Commands

1. `C-x C-c` - to save and quit Emacs 

2. `C-x C-s` - to save without quitting

3. `C-v` - move to next screen

4. `M-v` / `Esc-v` - move backwards to previous screen

5. ***Undo*** : `C-x u`

6. `C-/` : undoes the most recent change in the buffer.

7. `C-g`- keyboard-quit: if while typing a command you make a mistake and want to stop,this aborts a command in progress

11. <h3>Point, mark and region</h3>

Emacs uses the terms ***point, mark, and region*** to provide more precision about the selected text and position of the cursor.

- The ***point*** is the place in a buffer where editing (i.e. insertion) is currently taking place, and is usually indicated by a cursor.
- The ***mark*** is a marker placed anywhere in the buffer using commands like set-mark-command `C-SPC` or exchange-point-and-mark `C-x C-x`
- The ***region*** is the area between point and mark, and many commands operate on the region to e.g. delete, spell check, indent, or compile it.

To set a ***mark or mark your current location***, <br/>
Use the command `C-SPC` (***SPC is the space key***)

After doing this try moving the cursor to a different portion on the buffer.
You should notice some hightlighting.
Now wherever you cursor is type in the command `C-x C-x` to go back to initially marked location.

12. <h3> Searching </h3>

Emacs can do searches for strings (a ***string*** is a group of adjacent characters) either forward through the text or backward through it.
Searching for a string is a cursor motion command; it moves the cursor
to the next place where that string appears.

The command to initiate a search is `C-s` for ***forward search***, and `C-r`
for ***reverse search***.

NB: The Emacs search command is "incremental".  This means that the
search happens while you type in the string to search for.

When you type C-s you'll notice that the string ***I-search*** appears as
a prompt in the echo area.  This tells you that Emacs is in what is
called an ***incremental search*** waiting for you to type the thing that
you want to search for.  

<Return> and `C-g` terminates a search.
    
- If what you searched for has multiple occurences use the command `C-s` again to move througth these occurences.

- If you want to search for
something earlier in the text, type `C-r` instead.  Everything that we
have said about `C-s` also applies to `C-r`, except that the direction of
the search is ***reversed***. 
    
(`C-r` is basically used during an incremental search to move the cursor backward to find previous occurrences of the search text.)

13. <h3> Cut and Paste </h3>

In Emacs language, the process of cutting text is referred to as ***killing***, and the process of pasting text is known as ***yanking***.
    
Copying text, on the other hand, is referred to as ***saving onto the kill-ring***. This is because in Emacs, when you cut text, it is sent directly to the top of the kill-ring. The kill-ring is basically a list containing the previously killed (cut) blocks of text.
    
To copy or kill text in Emacs, you must first select the text. This is done by using the selection command `C + SPC`.


    
- ***To cut, or kill***, the text, you can use the keys `C-k` ***to kill a particular line***, or the `C-w` command ***to kill the entire selected region.***

- If you would only like to ***copy*** the selected region ***without removing it in the process***, this can be done by hitting `M-w`.

- ***To paste, or yank***, the text, press the keys `C-y`. This pastes the last killed item from the kill ring. 

By default, when you paste into the buffer, what gets pasted is the very last thing that you cut or copied. However, everything you have cut or copied is stored in the ***kill ring (which is analogous to a clipboard)***, and can be retrieved when you paste. Thus, you can paste earlier sections of cut or copied text. To do so, immediately after having pasted with `C-y`, press `M-y`. ***The text that was just pasted will be replaced by an earlier section of text you selected.*** You may ***press M-y multiple times to cycle backwards**** through all of your previously cut or copied text.

14. <h3> Running Commands by Name </h3>
    
Emacs has way too many keyboard shortcuts to memorize. Fortunately, ***you can call commands by name by typing `M-x` and the name of the command.***
    
So if you know the name of the command to execute, you can type it after `M-x`, and then ***press RET*** (the Return key, which is the same as the Enter key). For example, `M-x find-file` opens a file. 
`M-x save-buffer` saves the current file. You can use ***TAB to complete words***.We use < up > and < down > to go through your command history.

### Emacs vs. Vim ( After learning vim you can understand this portion better)

Emacs is basically a a (less popular) cousin of Vim.

One of the most notable differences between the two editors is that, ***unlike Emacs, Vim has two modes: Insert mode (where you can edit the file and cannot enter commands) and Command mode (where you can only enter commands and the file is read-only).*** 

***Because Emacs is modeless, its keyboard commands often start with the Ctrl key or the Meta key*** (which can be Esc or Opt if configured in your macOS terminal preferences), ***so that the system can distinguish actual edits from commands.*** In my experience, ***Emacs resembles editors like Microsoft Word and Google Docs*** more than Vim because of its modelessness, and this fact may make it easier to get used to than Vim.

A ***scratch buffer*** in Emacs is a temporary workspace or a blank canvas where you can quickly jot down notes, experiment with code, or perform various tasks without affecting any existing files or buffers. It serves as a scratchpad or a sandbox within Emacs.

When you start Emacs or open a new scratch buffer, you have a clean space to type and edit text. You can freely write, modify, and delete content in the scratch buffer without worrying about saving or preserving changes. It's like having a disposable piece of paper or a whiteboard that you can use for temporary work.

The scratch buffer is especially useful for trying out code snippets, testing small programs, or quickly drafting ideas. It allows you to experiment and explore without the need to create or manage separate files. Once you're done with your temporary work, you can either discard the scratch buffer or copy and paste the relevant parts into a more permanent location, such as a file or another buffer.

In summary, a scratch buffer in Emacs provides a convenient space for quick and temporary editing, note-taking, and experimentation without affecting existing files or buffers.

### Emacs 

Emacs is more than just a text editor; it can also issue shell commands, access the Internet, write and test programs, and read and send emails within the Emacs environment. 

Emacs can be further extended or customized using Emacs Lisp, a dialect of the Lisp programming language. Emacs Lisp includes numerous extensions such as a project planner, calendar, debugger interface, and a mail and news reader.

Another useful feature that renders text and code more readable is automatic indentation. Emacs identifies blocks of code and groups them accordingly. It also provides Unicode character support for practically all writing systems and languages, and features self-documentation, which automatically generates and displays documentation for every command, variable and internal function found in a program’s source code. </br>
